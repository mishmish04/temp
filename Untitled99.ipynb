````python
import pandas as pd
import requests
from time import sleep
import warnings
warnings.filterwarnings('ignore')

print("Fetching phishing URLs from OpenPhish...")
url = "https://openphish.com/feed.txt"
response = requests.get(url)
phishing_urls = response.text.strip().split('\n')[:300]  # Get 300 URLs

df_phish = pd.DataFrame({'url': phishing_urls, 'label': 'phishing'})
print(f"✓ Got {len(df_phish)} phishing URLs")

print("\nFetching legitimate URLs from Tranco...")
tranco_url = "https://tranco-list.eu/top-1m.csv.zip"
df_legit = pd.read_csv(tranco_url, names=['rank', 'url']).head(300)
df_legit['url'] = 'https://' + df_legit['url']
df_legit['label'] = 'legitimate'
df_legit = df_legit[['url', 'label']]
print(f"✓ Got {len(df_legit)} legitimate URLs")

df_combined = pd.concat([df_phish, df_legit], ignore_index=True)
print(f"\n✓ Total URLs: {len(df_combined)}")

df_combined.to_csv('urls_list.csv', index=False)
print("✓ URLs saved to 'urls_list.csv'")

def fetch_html(url, timeout=10):
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
        response = requests.get(url, headers=headers, timeout=timeout, verify=False, allow_redirects=True)
        return response.text if response.status_code == 200 else None
    except:
        return None

print("\nStarting to fetch HTML content...")
df_combined['html_content'] = None

for idx, row in df_combined.iterrows():
    print(f"Fetching {idx+1}/{len(df_combined)}: {row['url'][:60]}...")
    df_combined.at[idx, 'html_content'] = fetch_html(row['url'])
    
    # Save checkpoint every 50 URLs (so you don't lose progress!)
    if (idx + 1) % 50 == 0:
        df_combined.to_csv('dataset_checkpoint.csv', index=False)
        print(f"  ✓✓ Checkpoint saved at {idx+1} URLs\n")
    
    sleep(0.01)  

df_combined.to_csv('phishing_dataset_complete.csv', index=False)
print("\n" + "="*60)
print("✓✓✓ COMPLETE! Dataset saved to 'phishing_dataset_complete.csv'")
print("="*60)

df_clean = df_combined[df_combined['html_content'].notna()]
print(f"\nSuccessfully fetched HTML for {len(df_clean)} URLs")
print(f"  - Phishing: {len(df_clean[df_clean['label']=='phishing'])}")
print(f"  - Legitimate: {len(df_clean[df_clean['label']=='legitimate'])}")
````

````python
def url_to_sequence(url, char_to_idx, max_len=200):
    sequence = []
    for char in url[:max_len]:
        sequence.append(char_to_idx.get(char, char_to_idx['<UNK>']))
    pass
    while(len(sequence)) < max_len:
        sequence.append(0)
    return sequence
````

````python
from bs4 import BeautifulSoup

def extract_text_from_html(html):
    """Extract visible text from HTML"""
    try:
        soup = BeautifulSoup(html, 'html.parser')
        
        # Remove script and style elements
        for script in soup(['script', 'style']):
            script.decompose()
        #The decompose function destroys the unwanted tags like the script and the style, so that we can successfully extract the useful info
        
        text = soup.get_text(separator=' ')
        
        text = ' '.join(text.split())
        
        return text.lower()  
    except:
        return ""

df_clean['html_text'] = df_clean['html_content'].apply(extract_text_from_html)

# Check results
print(f"Sample HTML text (first 200 chars):")
print(df_clean['html_text'].iloc[0][:200])
````

````python
from collections import Counter

# Count all words across all documents
all_words = []
for words in df_clean['html_words']:
    all_words.extend(words)

word_counter = Counter(all_words)
print(f"Total unique words: {len(word_counter)}")
print(f"Most common words: {word_counter.most_common(20)}")

# Keep only top N most frequent words
VOCAB_SIZE = 5000  # Adjust this
most_common_words = [word for word, count in word_counter.most_common(VOCAB_SIZE)]

# Create vocab with special tokens
PAD_TOKEN = '<PAD>'
UNK_TOKEN = '<UNK>'

word_vocab = [PAD_TOKEN, UNK_TOKEN] + most_common_words

word_to_idx = {word: idx for idx, word in enumerate(word_vocab)}
idx_to_word = {idx: word for word, idx in word_to_idx.items()}

print(f"\nFinal word vocab size: {len(word_vocab)}")
````

````python
def html_to_sequence(words_list, word_to_idx, max_len=500):
    """
    Convert list of words to padded integer sequence
    
    Args:
        words_list: list of words (e.g., ['login', 'account', 'verify'])
        word_to_idx: dictionary mapping words to integers
        max_len: maximum sequence length
    
    Returns:
        list of integers (length = max_len)
    """
    sequence = []
    
    # Convert words to indices (truncate at max_len)
    for word in words_list[:max_len]:
        sequence.append(word_to_idx.get(word, word_to_idx['<UNK>']))
    
    # Pad with zeros if needed
    sequence.extend([0] * (max_len - len(sequence)))
    
    return sequence
````

````python
import ssl
import socket
from urllib.parse import urlparse
import hashlib
from datetime import datetime

def extract_ssl_features(url):
    """
    Extract SSL certificate features using cryptographic verification
    Returns dict with crypto-related features
    """
    features = {
        'has_https': 0,
        'has_valid_ssl': 0,
        'cert_trusted_ca': 0,
        'cert_age_days': -1,
        'url_hash': ''
    }
    
    try:
        parsed = urlparse(url)
        
        # Feature 1: HTTPS check
        features['has_https'] = 1 if parsed.scheme == 'https' else 0
        
        # Feature 2: URL hash (SHA-256 - cryptographic hash)
        features['url_hash'] = hashlib.sha256(url.encode()).hexdigest()
        
        # Feature 3-5: SSL Certificate validation (uses RSA/digital signatures)
        if parsed.scheme == 'https':
            hostname = parsed.netloc
            context = ssl.create_default_context()
            
            with socket.create_connection((hostname, 443), timeout=5) as sock:
                with context.wrap_socket(sock, server_hostname=hostname) as ssock:
                    cert = ssock.getpeercert()
                    
                    # Valid SSL certificate exists
                    features['has_valid_ssl'] = 1
                    
                    # Check if issued by trusted CA (digital signature verification)
                    issuer = dict(x[0] for x in cert['issuer'])
                    trusted_cas = ['DigiCert', 'Let\'s Encrypt', 'GlobalSign', 'Comodo', 'GoDaddy']
                    features['cert_trusted_ca'] = 1 if any(ca in issuer.get('organizationName', '') for ca in trusted_cas) else 0
                    
                    # Certificate age (newer certs on phishing sites)
                    not_before = datetime.strptime(cert['notBefore'], '%b %d %H:%M:%S %Y %Z')
                    cert_age = (datetime.now() - not_before).days
                    features['cert_age_days'] = cert_age
                    
    except:
        pass  # Use default values if SSL check fails
    
    return features

# Apply to all URLs
print("Extracting SSL/cryptographic features...")
ssl_features = df_clean['url'].apply(extract_ssl_features)
ssl_features_df = pd.DataFrame(ssl_features.tolist())

# Add to main dataframe
df_clean = pd.concat([df_clean, ssl_features_df], axis=1)

print("\n✓ Cryptographic features extracted!")
print(f"Features added: {ssl_features_df.columns.tolist()}")
print(f"\nFeature distribution:")
print(ssl_features_df[['has_https', 'has_valid_ssl', 'cert_trusted_ca']].sum())


````

````python
import torch
from torch.utils.data import Dataset, DataLoader

class PhishingDataset(Dataset):
    def __init__(self, X_url, X_html, y):
        self.X_url = torch.LongTensor(X_url)
        self.X_html = torch.LongTensor(X_html)
        self.y = torch.FloatTensor(y)
    
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X_url[idx], self.X_html[idx], self.y[idx]

train_dataset = PhishingDataset(X_url_train, X_html_train, y_train)
val_dataset = PhishingDataset(X_url_val, X_html_val, y_val)
test_dataset = PhishingDataset(X_url_test, X_html_test, y_test)

print(f"✓ Train dataset: {len(train_dataset)} samples")
print(f"✓ Val dataset: {len(val_dataset)} samples")
print(f"✓ Test dataset: {len(test_dataset)} samples")
````

````python
BATCH_SIZE = 32

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"✓ Train batches: {len(train_loader)}")
print(f"✓ Val batches: {len(val_loader)}")
print(f"✓ Test batches: {len(test_loader)}")

url_batch, html_batch, label_batch = next(iter(train_loader))
print(f"\nSample batch shapes:")
print(f"  URL: {url_batch.shape}")
print(f"  HTML: {html_batch.shape}")
print(f"  Labels: {label_batch.shape}")
````

````python
import torch.nn as nn
import torch.nn.functional as F

class WebPhishLite(nn.Module):
    def __init__(self, char_vocab_size=52, word_vocab_size=5002, 
                 char_emb_dim=16, word_emb_dim=16, url_len=200, html_len=500):
        super(WebPhishLite, self).__init__()
        
        self.url_embedding = nn.Embedding(char_vocab_size, char_emb_dim, padding_idx=0)
        self.html_embedding = nn.Embedding(word_vocab_size, word_emb_dim, padding_idx=0)
        
        self.conv3 = nn.Conv1d(char_emb_dim + word_emb_dim, 32, kernel_size=3, padding=1)
        self.conv5 = nn.Conv1d(char_emb_dim + word_emb_dim, 32, kernel_size=5, padding=2)
        self.conv7 = nn.Conv1d(char_emb_dim + word_emb_dim, 32, kernel_size=7, padding=3)
        
        self.pool = nn.MaxPool1d(kernel_size=2)
        self.layer_norm = nn.LayerNorm(96)
        
        self.fc1 = nn.Linear(96, 128)
        self.dropout1 = nn.Dropout(0.3)
        
        self.fc2 = nn.Linear(128, 64)
        self.dropout2 = nn.Dropout(0.3)
        
        self.fc3 = nn.Linear(64, 1)
    
    def forward(self, url, html):
        url_emb = self.url_embedding(url)
        html_emb = self.html_embedding(html)
        
        url_padded = F.pad(url_emb, (0, 0, 0, html.size(1) - url.size(1)))
        
        x = torch.cat([url_padded, html_emb], dim=2)
        x = x.permute(0, 2, 1)
        
        c3 = F.gelu(self.conv3(x))
        c5 = F.gelu(self.conv5(x))
        c7 = F.gelu(self.conv7(x))
        
        x = torch.cat([c3, c5, c7], dim=1)
        x = self.pool(x)
        
        x = x.permute(0, 2, 1)
        x = self.layer_norm(x)
        
        x = torch.mean(x, dim=1)
        
        x = F.gelu(self.fc1(x))
        x = self.dropout1(x)
        
        x = F.gelu(self.fc2(x))
        x = self.dropout2(x)
        
        x = self.fc3(x)
        
        return x.squeeze()

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

model = WebPhishLite().to(device)
print(f"✓ Model created with {sum(p.numel() for p in model.parameters())} parameters")
````

````python
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=3, factor=0.5)

print("✓ Loss function: BCEWithLogitsLoss")
print("✓ Optimizer: Adam (lr=0.001)")
print("✓ Scheduler: ReduceLROnPlateau")
````

````python
from sklearn.metrics import accuracy_score
import time

def train_epoch(model, loader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    all_preds = []
    all_labels = []
    
    for url, html, labels in loader:
        url, html, labels = url.to(device), html.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(url, html)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        preds = (torch.sigmoid(outputs) > 0.5).float()
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
    
    avg_loss = total_loss / len(loader)
    accuracy = accuracy_score(all_labels, all_preds)
    return avg_loss, accuracy

def validate(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for url, html, labels in loader:
            url, html, labels = url.to(device), html.to(device), labels.to(device)
            
            outputs = model(url, html)
            loss = criterion(outputs, labels)
            
            total_loss += loss.item()
            preds = (torch.sigmoid(outputs) > 0.5).float()
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    avg_loss = total_loss / len(loader)
    accuracy = accuracy_score(all_labels, all_preds)
    return avg_loss, accuracy

print("✓ Training and validation functions defined")
````

````python
NUM_EPOCHS = 30
best_val_loss = float('inf')
patience = 5
patience_counter = 0

train_losses = []
val_losses = []
train_accs = []
val_accs = []

print("Starting training...\n")

for epoch in range(NUM_EPOCHS):
    start_time = time.time()
    
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = validate(model, val_loader, criterion, device)
    
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    
    scheduler.step(val_loss)
    
    epoch_time = time.time() - start_time
    
    print(f"Epoch {epoch+1}/{NUM_EPOCHS} ({epoch_time:.1f}s)")
    print(f"  Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}")
    print(f"  Val Loss:   {val_loss:.4f}, Val Acc:   {val_acc:.4f}")
    
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), 'best_model.pth')
        print(f"  ✓ Best model saved!")
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"\nEarly stopping triggered at epoch {epoch+1}")
            break
    print()

print("✓ Training complete!")
````

````python
import requests
from bs4 import BeautifulSoup

def predict_url(url, model, char_to_idx, word_to_idx, device):
    """
    Predict if a URL is phishing
    """
    try:
        response = requests.get(url, timeout=5, verify=False)
        html = response.text
    except:
        html = ""
    
    soup = BeautifulSoup(html, 'html.parser')
    for script in soup(['script', 'style']):
        script.decompose()
    html_text = soup.get_text(separator=' ')
    html_text = ' '.join(html_text.split()).lower()
    
    url_lower = url.lower()
    url_seq = []
    for char in url_lower[:200]:
        url_seq.append(char_to_idx.get(char, char_to_idx['<UNK>']))
    url_seq.extend([0] * (200 - len(url_seq)))
    
    html_words = html_text.split()[:500]
    html_seq = []
    for word in html_words:
        html_seq.append(word_to_idx.get(word, word_to_idx['<UNK>']))
    html_seq.extend([0] * (500 - len(html_seq)))
    
    url_tensor = torch.LongTensor([url_seq]).to(device)
    html_tensor = torch.LongTensor([html_seq]).to(device)
    
    model.eval()
    with torch.no_grad():
        output = model(url_tensor, html_tensor)
        prob = torch.sigmoid(output).item()
    
    return prob, "Phishing" if prob > 0.5 else "Legitimate"

print("✓ Inference function defined")
````

# Security Analysis: WebPhish-Lite++

## 1. Threat Model

### Attacker Capabilities
- **Typosquatting**: Registering domains similar to legitimate ones (e.g., `g00gle.com` vs `google.com`)
- **Homoglyphs**: Using unicode characters that look identical (e.g., `аpple.com` with Cyrillic 'а')
- **URL Obfuscation**: Using IP addresses, URL shorteners, or encoded characters
- **HTML Mimicry**: Copying legitimate website HTML/CSS to create convincing fakes
- **SSL Certificate Abuse**: Obtaining valid SSL certificates for phishing domains (Let's Encrypt)

### Attack Vectors
1. Email phishing with embedded links
2. SMS phishing (smishing)
3. Social media message phishing
4. Compromised websites with injected phishing content
5. Search engine poisoning

## 2. Cryptographic Concepts in Detection

### 2.1 Digital Signatures & SSL/TLS
Our model extracts SSL certificate features:
- **RSA/ECDSA Signatures**: Verifies if certificate is signed by trusted CA
- **Certificate Age**: Phishing sites often use newly issued certificates
- **Trust Chain**: Validates certificate authority hierarchy

```python

```

### 2.2 Hashing (SHA-256)
URL hashing for unique identification:
```python
url_hash = hashlib.sha256(url.encode()).hexdigest()
```
- Provides cryptographic fingerprint of URL
- Detects exact duplicate phishing URLs
- Collision-resistant (SHA-256 properties)

### 2.3 TLS Encryption
Detection process verifies:
- HTTPS presence (transport layer security)
- Valid TLS handshake completion
- Certificate chain verification

## 3. RSA & ElGamal Context

### RSA in SSL/TLS
- **Key Exchange**: RSA used in TLS 1.2 for key exchange
- **Digital Signatures**: RSA signs SSL certificates
- **Public Key Infrastructure**: CA uses RSA to sign certificates

### ElGamal Applicability
- **Alternative to RSA**: Can be used for digital signatures
- **Elliptic Curve Variant**: ECDSA (used in modern SSL) based on ElGamal principles
- **Security Properties**: Semantic security similar to RSA

## 4. Model Security Features

### Input Sanitization
- URL length truncation (prevents buffer overflow-style attacks)
- HTML tag stripping (prevents script injection)
- Character encoding normalization

### Robustness Against Adversarial Examples
- Character-level embeddings resistant to typos
- Multi-kernel convolutions detect patterns at different scales
- Dropout prevents overfitting to specific phishing patterns

## 5. Deployment Security

### API Security
- **Rate Limiting**: Prevent model probing attacks
- **Input Validation**: Sanitize URLs before processing
- **Sandboxed HTML Fetching**: Isolate network requests
- **HTTPS Only**: Encrypt API communications

### Model Integrity
```python

model_hash = hashlib.sha256(model_bytes).hexdigest()
assert model_hash == EXPECTED_HASH
```

### Threat Mitigations
| Threat | Mitigation |
|--------|-----------|
| Model Extraction | Rate limiting, API authentication |
| Adversarial URLs | Ensemble predictions, confidence thresholds |
| Data Poisoning | Continuous retraining with verified labels |
| DoS Attacks | Request throttling, timeout limits |

## 6. Certificate Inspection Pipeline

```python
# Certificate validation process
1. Extract certificate from TLS handshake
2. Verify signature using CA public key (RSA/ECDSA)
3. Check validity period (not expired)
4. Validate domain name match
5. Check revocation status (OCSP/CRL)
```

## 7. Future Security Enhancements

1. **Homomorphic Encryption**: Allow encrypted model inference
2. **Federated Learning**: Train on distributed data without sharing
3. **Differential Privacy**: Add noise to prevent training data leakage
4. **Blockchain Verification**: Immutable phishing URL database
5. **Zero-Knowledge Proofs**: Verify URL without revealing it
